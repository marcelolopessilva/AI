### Adicionando stopwords específicas após análise da contagem de palavras

```python
nltk_pt_stopwords.append('nº')
nltk_pt_stopwords.append('n°')
nltk_pt_stopwords.append('n° 583')
nltk_pt_stopwords.append('ii')
nltk_pt_stopwords.append('iii')
nltk_pt_stopwords.append('ser')
nltk_pt_stopwords.append('ser')
nltk_pt_stopwords.append('§')
nltk_pt_stopwords.append('mp')
nltk_pt_stopwords.append('003 10')
nltk_pt_stopwords.append('publicado 04')
```

### Limpeza de ruído - Transformar em minúsculo, limpeza de caracteres especiais

```python
import regex as re
import string

def limpa_ruido_texto(texto):
    texto = texto.lower()
    texto = re.sub('[%s]' % re.escape(string.punctuation), '', texto) # Retira pontuações
    texto = re.sub('\w*\d\w*', '', texto) # Retira números entre palavras
    texto = re.sub('§–', '', texto) # Casos específicos para os textos do DOU
    return texto

base_total['informacao'] = base_total['informacao'].apply(lambda x: limpa_ruido_texto(x))

base_total['informacao'].head()
```

### Quantidade de stop words na base

```python
import matplotlib.pyplot as plt
%matplotlib inline

def grafico_top_stopwords(texto, titulo):
    
    nova= texto.str.split()
    nova=nova.values.tolist()

    corpus=[palavra for i in nova for palavra in i]
    from collections import defaultdict
    dic=defaultdict(int)
    for palavra in corpus:
        if palavra in nltk_pt_stopwords:
            dic[palavra]+=1
            
    top=sorted(dic.items(), key=lambda x:x[1],reverse=True)[:10]
    x,y=zip(*top)
    plt.figure(figsize=(10,10))
    plt.bar(x,y)
    plt.title(titulo)

grafico_top_stopwords(base_total['informacao'], "Quantidade de stop words na base")
```

```python
# Removendo stopwords
import nltk

from nltk.tokenize import word_tokenize

# Tokenizando a informação
base_total['informacao_token']  = base_total['informacao'].apply(word_tokenize)

base_total['informacao_token'] = base_total['informacao_token'].apply(lambda x: [palavra for palavra in x if palavra not in nltk_pt_stopwords])

base_total['informacao'] = [' '.join(map(str, l)) for l in base_total['informacao_token']]

base_total.head()
```

### Top palavras que não são stopwords

```python
import seaborn as sns
from collections import  Counter

def grafico_top_non_stopwords(texto, titulo):
   
    novo=texto.str.split()
    novo=novo.values.tolist()
    corpus=[palavra for i in novo for palavra in i]

    counter=Counter(corpus)
    mais_comum=counter.most_common()
    x, y=[], []
    for palavra,count in mais_comum[:30]:
        if (palavra not in nltk_pt_stopwords):
            x.append(palavra)
            y.append(count)
    plt.figure(figsize=(10,10))
    plt.title(titulo)
    sns.barplot(x=y,y=x)
```

```python
grafico_top_non_stopwords(base_total['informacao'], "Top 30 não stopwords")
```

### Distribuição percentual por classe

```python
plt.figure(figsize=(14,18))

nshow=30

idx = (base_total['interesse'] == 1).values
conta_palavras = np.asarray(text_features.mean(axis=0))[0]
conta_palavras_si = np.asarray(text_features[~idx].mean(axis=0))[0]
conta_palavras_ci = np.asarray(text_features[ idx].mean(axis=0))[0]
idx = np.argsort(conta_palavras)[-nshow:]
vocabulary = np.array(list(vectorizer.vocabulary_.keys()))[idx].tolist()
yaxis = np.arange(nshow)

total_count = conta_palavras_si + conta_palavras_ci

bar_neg = plt.barh(yaxis, conta_palavras_si[idx] / total_count[idx], label='Sem Interesse', color='#ADD8E6')
bar_pos = plt.barh(yaxis, conta_palavras_ci[idx] / total_count[idx],
                   left=conta_palavras_si[idx] / total_count[idx], label='Interesse', color='#FF8C00')

plt.plot([0.5, 0.5],[-0.5, nshow], '-k', label='Referência - 50%')
plt.yticks(yaxis, vocabulary)
plt.legend()
plt.ylim((-0.5, nshow-0.5))

plt.title('Proporção Por Classe')
```

### Distribuição do TfIdf Médio por Classe

```python
plt.figure(figsize=(14,18))

nshow=30

idx = (base_total['interesse'] == 1).values
media_palavra     = np.asarray(text_features.mean(axis=0))[0]
media_palavra_si = np.asarray(text_features[~idx].mean(axis=0))[0]
media_palavra_ci = np.asarray(text_features[ idx].mean(axis=0))[0]
idx = np.argsort(media_palavra)[-nshow:]
vocabulary = np.array(list(vectorizer.vocabulary_.keys()))[idx].tolist()
yaxis = np.arange(nshow)

bar_neg = plt.barh(yaxis, media_palavra_si[idx], label='Sem Interesse')
bar_pos = plt.barh(yaxis, media_palavra_ci[idx], left=media_palavra_si[idx], label='Interesse', color='red')
plt.yticks(yaxis, vocabulary)

plt.legend()
plt.title('Proporção Por Classe')
plt.xlabel('TfIdf médio')
```
