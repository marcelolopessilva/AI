## <span style="color:#DC143C">**Biblioteca MISSINGNO**

<b>Missingno</b> é uma biblioteca que permite que você visualize dados nulos em seu dataset.
<p>    
Na biblioteca missingno, existem quatro tipos de gráficos para visualizar a integridade dos dados: o gráfico de barras, o gráfico de matriz, o mapa de calor e o gráfico de dendrograma. Cada um tem suas próprias vantagens para identificar dados ausentes.

### <span style="color:#228B22">**Importar os dados**

```python
# Carga do arquivo csv preparada para a análise exploratória de dados. Ajustes de separador, primeira coluna como índice e parse de Data.
df_dados = pd.read_csv('Dados_missingno.csv', delimiter=';', index_col=0)
```

### <span style="color:#CD5C5C">**MISSINGNO.Bar plot**

O gráfico de barras fornece um gráfico simples em que cada barra representa uma coluna dentro do dataframe. 
A altura da barra indica o quão completa essa coluna é, ou seja, quantos valores não nulos estão presentes. 
<br>
Ele pode ser gerado chamando: <b>msno.bar(df)</b>
<p>
No lado direito do gráfico, a escala é medida em valores de índice. Com o canto superior direito representando o número máximo de linhas no dataframe.<br>
Na parte superior do gráfico, há uma série de números que representam a contagem total dos valores não nulos nessa coluna.

```python
msno.bar(df_dados)
```

### <span style="color:#CD5C5C">**MISSINGNO.Matrix Plot**

O gráfico de matriz é uma ótima ferramenta se você estiver trabalhando com dados relacionados à profundidade ou dados de série temporal. 
Ele fornece um preenchimento de cor para cada coluna. Quando os dados estão presentes, o gráfico é sombreado em cinza (ou a cor de sua escolha) e, 
quando está ausente, o gráfico é exibido em branco.
<br>
O gráfico de matriz pode ser gerado chamando: <b>msno.matrix(df)</b>
<p>
No lado direito do gráfico está um minigráfico que varia de 0 à esquerda ao número total de colunas no dataframe à direita. Um close pode ser visto acima. Quando uma linha tem um valor em cada coluna, a linha estará na posição máxima certa. Conforme os valores ausentes começam a aumentar dentro dessa linha, a linha se moverá para a esquerda.

```python
msno.matrix(df_dados)
```

### <span style="color:#CD5C5C">**MISSINGNO.heatmap Plot**

O mapa de calor é usado para identificar correlações de nulidade entre cada uma das diferentes colunas. Em outras palavras, pode ser usado para identificar se existe uma relação na presença de valores nulos entre cada uma das colunas.
<p>
Valores <b>próximos de 1 positivo</b> indicam que a presença de valores nulos em uma coluna está correlacionada com a presença de valores nulos em outra coluna.<br>
Valores <b>próximos de 1 negativo</b> indicam que a presença de valores nulos em uma coluna é anticorrelacionada com a presença de valores nulos em outra coluna. Em outras palavras, quando valores nulos estão presentes em uma coluna, há valores de dados presentes na outra coluna e vice-versa.<br>
Valores <b>próximos de 0</b> indicam que há pouca ou nenhuma relação entre a presença de valores nulos em uma coluna em comparação com outra.
<p>
O mapa de calor pode ser gerado pelo seguinte código: <b>msno.heatmap(df)</b>
<p>
A abordagem do mapa de calor é mais adequada para conjuntos de dados menores.

```python
msno.heatmap(df_dados)
```

### <span style="color:#CD5C5C">**MISSINGNO.dendrogram**

O gráfico de dendrograma fornece um gráfico semelhante a uma árvore gerado por agrupamento hierárquico e agrupa colunas que 
possuem fortes correlações em nulidade.<br>
Se várias colunas forem agrupadas no nível zero, a presença de nulos em uma dessas colunas está diretamente relacionada à presença ou 
ausência de nulos nas outras colunas. Quanto mais separadas as colunas na árvore, menos provável que os valores nulos possam ser correlacionados 
entre as colunas.

O dendrograma pode ser gerado por: <b>msno.dendrogram(df)</b>

```python
msno.dendrogram(df_dados)
```

### <span style="color:#228B22">**Exportar Grafico para Imagem (externa)**

```python
export_grafico = msno.matrix(df_dados)
export_grafico.get_figure().savefig("grafico_barplot.png")
```
