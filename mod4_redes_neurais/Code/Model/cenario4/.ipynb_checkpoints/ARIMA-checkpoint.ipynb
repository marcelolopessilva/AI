<h3>O QUE Ã‰ O ALGORITMO HOLT-WINTERS E COMO FUNCIONA?</h3>

Esse algoritmo Ã© muito utilizado quando hÃ¡ a necessidade de calcular previsÃµes em sÃ©ries temporais que possuam tendÃªncia e sazonalidade.

O algoritmo Holt-Winters Ã© uma das tÃ©cnicas de previsÃ£o mais populares para sÃ©ries temporais. Apesar de existir hÃ¡ dÃ©cadas, ele ainda Ã© muito utilizado em aplicativos voltados para fins de detecÃ§Ã£o de anomalias e, especialmente, na previsÃ£o de tempo.

Esse modelo Ã© uma extensÃ£o do modelo de Holt (suavizaÃ§Ã£o exponencial dupla), desenvolvido por Winter. A sua capacidade de previsÃ£o Ã© simples, mas muito poderosa. Ele pode lidar com muitos padrÃµes sazonais complicados, simplesmente encontrando o valor central e adicionando os efeitos de inclinaÃ§Ã£o e sazonalidade.

Vantagem:
1. Simples implementaÃ§Ã£o
SÃ£o necessÃ¡rias apenas 3 informaÃ§Ãµes para a sua aplicaÃ§Ã£o: previsÃ£o para o perÃ­odo mais recente, valor real para esse perÃ­odo e o valor da constante de suavizaÃ§Ã£o, um fator de ponderaÃ§Ã£o que reflete o peso concedido aos valores de dados mais recentes.

2. Realiza previsÃµes precisas
Um mÃ©todo de suavizaÃ§Ã£o exponencial produz uma previsÃ£o para um perÃ­odo futuro. Ao se utilizar a tÃ©cnica de projeÃ§Ã£o de tendÃªncia, previsÃµes futuras poderÃ£o ser geradas. Ela Ã© considerada precisa por levar em consideraÃ§Ã£o a diferenÃ§a entre as projeÃ§Ãµes reais e o que realmente aconteceu.

3. Confere maior importÃ¢ncia a observaÃ§Ãµes recentes
Os dados observados representam a soma de dois ou mais componentes. Um deles Ã© o erro aleatÃ³rio, ou seja, a diferenÃ§a entre o valor observado e o valor verdadeiro. Assim, a variaÃ§Ã£o aleatÃ³ria Ã© negligenciada, tornando muito mais fÃ¡cil perceber o fenÃ´meno subjacente.

```python
count = int(data.shape[0]*0.6) # Treinei com 60%
Train = data[:count] # => 5227 linhas
Test = data[count:] # => 3485 linhas

y_hat_avg = Test.copy()

fit1 = Holt(np.asarray(Train['Vento'])).fit()

y_hat_avg['Holt_Winter'] = fit1.predict(start=count+1,end=data.shape[0])

plt.figure(figsize=(16,8))
plt.plot(Train.index, Train['Vento'], label='Train')
plt.plot(Test.index,Test['Vento'], label='Test')
plt.plot(y_hat_avg.index,y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
```

```python
history = [x for x in Train]
history
```

```python
import statsmodels.tsa.arima.model as ARIMA

predictions = []
lower_list = []
upper_list = []

for t in range(len(Test)):
   
    model = ARIMA()   
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    lower = output[2][0][0]
    upper = output[2][0][1]
    
    predictions.append(yhat)
    lower_list.append(lower)
    upper_list.append(upper)
    obs = Test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
    
#print(model_fit.summary())    
```

```python
error = metrics.mean_squared_error(Test, predictions)
print('Test MSE: %.3f' % error)
```

```python
error = metrics.mse(Test, predictions)
print('Test MSE: %.3f' % error)
```

```python
# plot
plt.plot(Test.values,color='black')
plt.plot(lower_list,color='red')
plt.plot(upper_list,color='green')
plt.plot(predictions)
plt.show()
```
