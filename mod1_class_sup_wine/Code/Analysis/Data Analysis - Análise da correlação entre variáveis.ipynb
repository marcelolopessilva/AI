```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.precision", 4)

df_wine = pd.read_parquet("../../Data/Processed/wine/df_wine.parquet") 
```

```python
# Correlação entre as variáveis

df_wine.corr()
```

```python
# Scatter Plot

# Referência: https://seaborn.pydata.org/generated/seaborn.pairplot.html'''

import seaborn as sb
sb.pairplot(df_wine)
```

```python
# Scatterplot com kde e regressão linear nas correlações

sb.pairplot(df_wine, diag_kind="kde", kind="reg")
```

***Pelos gráficos apresentados podemos afirmar que a maioria das variáveis não apresenta alta correlação, de forma que não seria possível a redução da quantidade de variáveis nestes casos. A direção destas relações também são variadas dependendo das correlações. A confirmação de quanto essas correlações são significantes pode ser melhor avaliada observando-se o R<sup>2</sup> de cada uma.***

***O diagnóstico utilizado por KDEs (Kernel Density Estimation) indicam as Funções de Probabilidade de Densidade (PDFs) de cada variável. Podemos observar PDFs esguias com densidade de probabilidade pequena e alongada (2a Na diagonal de cima para baixo). PDFs com múltiplos picos (2a Na diagonal de baixo para cima), PDFs com distribuições quase normais (5a na diagonal de baixo para cima) e outras. Todas passíveis de normalização por métodos logarítmos, mínimos e máximos e otros, de forma que a estabilidade dos modelos sejam garantidas.***
