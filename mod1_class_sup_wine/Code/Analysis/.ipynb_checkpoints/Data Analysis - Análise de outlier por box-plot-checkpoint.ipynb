```python
import pandas as pd
import matplotlib. pyplot as plt
    
pd.set_option("display.precision", 4)

df_wine = pd.read_parquet("../../Data/Processed/wine/df_wine.parquet") 
```

```python
#Box Plot
#Referência: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html

for (column_name, _ ) in df_wine.items():
    plt.figure()
    df_wine.boxplot(column = column_name)
    plt.title(column_name)
```

***Notamos que algumas váriáveis apresentam muitos outliers tanto acima do limite superior, quanto do limite inferior, como fixed acidity, citric acid, pH e outros. Praticamente todas as variáveis apresentaram outliers. Devemos avaliar se esses outliers devem ser excluídos da análise, se devemos substituir seus valores por outros, por exemplo a média dos quartis Q3 para superiores e Q1 para inferiores ou alguma transformação que reduza a sua importância. Isso não se aplica, claro, se o objetivo da análise é justmente descobrir os outliers.***
