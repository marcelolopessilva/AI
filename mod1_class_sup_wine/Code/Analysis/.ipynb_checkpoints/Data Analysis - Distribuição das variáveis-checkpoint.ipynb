```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.precision", 4)

df_wine = pd.read_parquet("../../Data/Processed/wine/df_wine.parquet") 
```

```python
# Histogramas com as distribuições das variáveis'''

# Referência: https://matplotlib.org/gallery/index.html'''

for (column, _ ) in df_wine.items():
    plt.figure()
    df_wine[column].plot.hist(bins=50, figsize=(15,5), legend=True);
    plt.legend(ncol = 1)
    plt.title(column)
```

**Observamos vários padrões de distribuição de frequencia onde podem ser utilizadas técnicas de normalização para aumentar a estabilidade dos modelos escolhidos. Retirando as varáveis de qualidaade e tipo, podemos usar, por exemplo, a técnica de normalização logarítmica em free sulfor dioxide ou residual sugar que apresentam caudas.**

```python
# Utilizar KDE - Kernel Density Estimate

# Referência: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.kde.html

for (column, _ ) in df_wine.items():
    plt.figure()
    df_wine[column].plot.kde(figsize=(15,5), legend=True);
    plt.legend(ncol = 1)
    plt.title(column)
```
